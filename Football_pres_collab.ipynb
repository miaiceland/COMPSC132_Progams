```python
features=['Off.Rank_prev', 'Off.Plays_prev', 'Off.Yards_prev', 'Off.Yards.Play_prev', 'Off.TDs_prev', 'Off.Yards.per.Game_prev']
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#This combines data some data from 2 years to create train data
data17=df17.loc[:,['Team','Off.Rank', 'Off.Plays', 'Off.Yards', 'Off.Yards.Play', 'Off.TDs', 'Off.Yards.per.Game']]
data16=df16.loc[:,['Team','Off.Rank', 'Off.Plays', 'Off.Yards', 'Off.Yards.Play', 'Off.TDs', 'Off.Yards.per.Game']]

data_unclean=pd.merge(data17,data16, on='Team',
                  suffixes=('','_prev'), how='left')

data=data_unclean.dropna(axis=0)
```

```python
data
```

```python
#This fits the model
model=LinearRegression()
model.fit(data[features],data[target])
```

```python
#This combines data from 2 more different years to create test data
data19=df19.loc[:,['Team','Off.Rank', 'Off.Plays', 'Off.Yards', 'Off.Yards.Play', 'Off.TDs', 'Off.Yards.per.Game']]
data18=df18.loc[:,['Team','Off.Rank', 'Off.Plays', 'Off.Yards', 'Off.Yards.Play', 'Off.TDs', 'Off.Yards.per.Game']]

data_unclean_test=pd.merge(data19,data18, on='Team',
                  suffixes=('','_prev'), how='left')

data_test=data_unclean_test.dropna(axis=0)
```

```python
predictions=model.predict(data_test.loc[:,['Off.Rank_prev', 'Off.Plays_prev', 'Off.Yards_prev', 'Off.Yards.Play_prev',
                                           'Off.TDs_prev', 'Off.Yards.per.Game_prev']])
```

```python
predictions
```

```python
data_test.insert(6,'predictions', predictions)
data_test=data_test.set_index('Team')

print(data_test.loc[['Penn St. (Big Ten)']])
```

```python
touchdowns=data_test['Off.TDs'].to_list()
pred=data_test['predictions'] .to_list()

def check_accuracy(touchdowns, pred):
  diff=[]

  for i in range(len(pred)):
    diff.append(abs(touchdowns[i] - pred[i]))

  avg_diff=sum(diff) / len(diff)

  return avg_diff
```

```python
check_accuracy(touchdowns,pred)
```
